## Part (a)

For proving Abelian group, we need to prove the following properties:

### Closure

Consider two elements, $a$ and $b$ which are part of $R$\\{-1}, then:

$$a*b = ab + a + b$$

Since, $a$ and $b$ are real numbers, $ab$, $a$ and $b$ will also be real numbers, and thus, their sum will also be real. We only need to check if $a*b$ can result in -1.

Let's assume that for $a$ and $b$ not equal to -1, $a*b$ is -1. Then:

$$ab + a + b = -1$$

$$a(b+1) + b = -1$$

$$a(b+1) = -(b+1)$$

$$a = -1$$ 

The above step can be performed, since, $b \ne 1$

But, we know that $a \ne 1$, so, $ab + a + b \ne 1$.

Hence, we have proven the closure property.

### Associativity


Consider three elements, $a$, $b$ and $c$ which are part of $R$\\{-1}, then:

$$a*(b*c) = a*(bc + b + c)$$

$$a*(b*c) = a(bc + b + c) + a + (bc + b + c)$$

$$a*(b*c) = abc + ab + ac + a + bc + b + c$$

$$a*(b*c) = abc + ac + bc + c + ab + a + b$$

$$a*(b*c) = (ab + a + b)c + (ab + a + b) + c$$

$$a*(b*c) = (a*b)c + (a*b) + c$$

$$a*(b*c) = (a*b)*c$$

Hence, we have proven the associativity property.


### Neutral Element

Consider two elements, $a$ and $e$ which are part of $R$\\{-1}, such that:

$a*e = a$ and $e*a = a$

We need to check if such an element $e$ exists or not.

$$a*e = ae + a + e$$

$$a = ae + a + e$$

$$ae + e = 0$$

$$(a+1)e = 0$$

Since, $a \ne 1$

$$e = 0$$

We also know that {0} is a part of $R$\\{-1}, so, we have found that such an element exists. Now, let's verify if $e*a = a$ is true or not.

$$e*a = ea + a + e$$

$$a = ea + a + e$$

$$ea + e = 0$$

$$(a+1)e = 0$$

$$e = 0$$

We have reached at the same element.

Hence, we have proved that a neutral element exists.


### Inverse Element


Consider two elements, $a$ and $b$, and the neutral element, $e$, all lying in $R$\\{-1}. We have to check if the following condition can be fulfilled for any $b$ or not, given $a$.

$$a * b = e$$

$$ab + a + b = e$$

Since we know that $e = 0$

$$ab + a + b = 0$$

$$a(b+1) = -b$$

Since $b \ne 1$

$$a = \frac{-b}{b+1}$$

For a real number $b$, the above $a$ will also be a real number. We need to check if $a$ can be -1 or not. Let's assume that $a = -1$, then,

$$-1 = \frac{-b}{b+1}$$

$$b+1 = b$$

$$1 = 0$$

The above statement is clearly false, which means that $a \ne -1$. Thus, $a$ also lies in $R$\\{-1}

Hence, we have seen that for any $b$ lying in $R$\\{-1}, an inverse element $a = \frac{-b}{b+1}$ lying in $R$\\{-1} also exists.


### Commutativity

Consider two elements $a$ and $b$ lying in $R$\\{-1}, then

$$a * b = ab + a + b$$

$$a * b = ba + b + a$$

$$a * b = b * a$$

This proves the commutative property as well.

Since all the above properties hold true, we can $(R$\\{-1}, $*)$ is an Abelian group.
